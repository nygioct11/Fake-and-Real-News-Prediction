Import Necessary Libraries

```python
import pandas as pd
import numpy as np
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
import nltk
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
```

Loading the dataset

```python
true=pd.read_csv("/content/drive/MyDrive/fake and real news/true.csv",encoding="ISO-8859-1")
true
```

```python
fake=pd.read_csv("/content/drive/MyDrive/fake and real news/fake.csv",encoding="ISO-8859-1")
fake
```

```python
true["label"]=1
fake["label"]=0
```

```python
true.head()
```

```python
fake.head()
```

```python
true.isna().sum()
```

```python
fake.isna().sum()
```

```python
news=pd.concat([fake,true],axis=0)
news
```

```python
news.dtypes
```

```python
news=news.drop(["title","subject","date"],axis=1)
news
```

```python
news=news.sample(frac=1)
news
```

```python
news=news.reset_index()
```

```python
news.head()
```

```python
news.drop(["index"],axis=1,inplace=True)
```

```python
news
```

```python
text=news.text
text
```

```python
text=text.str.replace("[^a-zA-Z 0-9]","")
text
```

```python
stemmer=SnowballStemmer("english")
nltk.download('punkt')
```

```python
text=text.apply(lambda row:[stemmer.stem(token.lower())for token in word_tokenize(row)]).apply(lambda token:" ".join(token))
text
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
sw=stopwords.words("english")
text=text.apply(lambda row:[token for token in word_tokenize(row) if token not in sw]).apply(lambda token:" ".join(token))
text
```

```python
text=text.apply(lambda row:[token for token in word_tokenize(row) if len(token)>2]).apply(lambda token:" ".join(token))
text
```

```python
y=news["label"]
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
data=vectorizer.fit_transform(text)
print(data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data,y,test_size=0.3,random_state=1)
```

Model Training

```python
nb=MultinomialNB()
dt=DecisionTreeClassifier(criterion="entropy",random_state=1,max_depth=4)
rf=RandomForestClassifier(random_state=1)
ab=AdaBoostClassifier(random_state=1)
models=[nb,dt,rf,ab]
for model in models:
  print(model)
  model.fit(X_train,y_train)

  y_pred=model.predict(X_test)
  print(classification_report(y_test,y_pred))
```
